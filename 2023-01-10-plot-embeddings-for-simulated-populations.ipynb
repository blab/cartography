# Visualize simulated population embeddings for coronavirus-like populations with high recombination

```python
metadata = pd.read_csv(
    "simulations/coronavirus-like/high-recombination-rate/replicate-2/metadata.tsv",
    sep="\t",
    index_col=0,
)
```

```python
metadata.head()
```

```python
metadata["is_recombinant"].value_counts()
```

## PCA

```python
pca = pd.read_csv(
    "simulations/coronavirus-like/high-recombination-rate/replicate-2/gridsearch/embedding_components~nan/perplexity~nan/learning_rate~nan/method~pca/nearest_neighbors~nan/min_dist~nan.csv",
    index_col=0,
)
```

```python
pca = pca.join(metadata)
```

```python
pca.head()
```

```python
sns.pairplot(
    pca,
    x_vars=["pca1", "pca2", "pca3", "pca4", "pca5"],
    y_vars=["pca1", "pca2", "pca3", "pca4", "pca5"],
    hue="generation",
)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5), dpi=120)
ax.plot(
    pca.query("is_recombinant == False")["pca1"],
    pca.query("is_recombinant == False")["pca2"],
    "o",
    label="non-recombinant",
    alpha=0.2
)

ax.plot(
    pca.query("is_recombinant == True")["pca1"],
    pca.query("is_recombinant == True")["pca2"],
    "o",
    label="recombinant",
    alpha=0.75,
    zorder=10,
)

ax.set_xlabel("PC 1")
ax.set_ylabel("PC 2")

sns.despine()
```
